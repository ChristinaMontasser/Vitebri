```python
#dataContained --> Matrix contains states and indicators names
#pi --> Initial Matrix
#A --> #Transition Matrix
#B -->Emission Matrix
#Observation --> #Sequence of observations
#Return --> Squence of predicated states
def HMM(dataContained, pi, A, B, Observation):
    Max_Matrix=[0]* len(Observation)
    Mat1=[0]* len(Observation)
    Mat2=[0]* len(Observation)

    if(Observation[0]==dataContained[1][0]):
        Mat1[0] = pi[0]*B[0][0]
        Mat2[0]  = pi[1]*B[1][0]
        Max_Matrix[0]= dataContained[0][0] if Mat1[0] == max(Mat1[0], Mat2[0]) else dataContained[0][1] 
        #print(Mat1[0], Mat2[0])
    elif(Observation[0]==dataContained[1][1]):
        Mat1[0] = pi[0]*B[0][1]
        Mat2[0]  = pi[1]*B[1][1]
        Max_Matrix[0]= dataContained[0][0] if Mat1[0] == max(Mat1[0], Mat2[0]) else dataContained[0][1]  
        #print(Mat1[0], Mat2[0])
    for i in range(1, len(Observation), 1):
            if(Observation[i]==dataContained[1][0]): #D
                Mat1[i] = max(Mat1[i-1]*A[0][0]*B[0][0], Mat2[i-1]*A[1][0]*B[0][0])
                Mat2[i] = max(Mat1[i-1]*A[0][1]*B[1][0], Mat2[i-1]*A[1][1]*B[1][0])
                Max_Matrix[i]= dataContained[0][0] if Mat1[i] == max(Mat1[i], Mat2[i]) else dataContained[0][1]
                #print(Mat1[i], Mat2[i])
            elif(Observation[i]==dataContained[1][1]):
                Mat1[i] = max(Mat1[i-1]*A[0][0]*B[0][1], Mat2[i-1]*A[1][0]*B[0][1])
                Mat2[i] = max(Mat1[i-1]*A[0][1]*B[1][1], Mat2[i-1]*A[1][1]*B[1][1])
                Max_Matrix[i]= dataContained[0][0] if Mat1[i] == max(Mat1[i], Mat2[i]) else dataContained[0][1]
                #print(Mat1[i], Mat2[i])
    return Max_Matrix 
```

```python
dataContained1 = [['h', 'C'],['D', 'N']] #Matrix contains states and indicators names
pi1 = [0.55, 0.45] #initial Matrix
A1 = [ #Transion Matrix
        [0.75, 0.25], #H
        [0.3, 0.7]    #C
]

B1 = [ #Emission Matrix
     #D   #N
    [0.7, 0.3], #H
    [0.2, 0.8]  #C
]
Observation1 = ['N', 'N', 'N', 'D'] #Set of observations
HMM(dataContained1, pi1, A1, B1, Observation1)
```
